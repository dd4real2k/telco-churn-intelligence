```python
def clean_telco_data(df):
    # Create a copy so we don't overwrite the original data
    df_clean = df.copy()

    # Convert TotalCharges to numeric. 'coerce' turns those empty strings into NaN (nulls)
    df_clean['TotalCharges'] = pd.to_numeric(df_clean['TotalCharges'], errors='coerce')

    # Fill the new NaNs with 0 (since tenure is usually 0 for these rows)
    df_clean['TotalCharges'] = df_clean['TotalCharges'].fillna(0)

    return df_clean
```

```python
df_clean = clean_telco_data(df)
df_clean.head()
```

```python
# 1. Define the file path
CLEANED_DATA_PATH = 'telco_churn_cleaned.csv'

# 2. Define the function
def save_clean_data(df, path=CLEANED_DATA_PATH):
    df.to_csv(path, index=False)
    print("File successfully saved!")

save_clean_data(df_clean)
print(f"Cleaned dataset saved to: {CLEANED_DATA_PATH}")
```
